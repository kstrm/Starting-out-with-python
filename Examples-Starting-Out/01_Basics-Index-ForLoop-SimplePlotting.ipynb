```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format='retina' # hig-res plots for a Retina display 
```

```python
N = 10
x = np.linspace(1,10,N) # produces an array with N number of elements evenly spaced from 1 to 10
x
```

```python
x = np.zeros(N) # produces an array with N elements...floating point
x
```

```python
x = np.arange(N) # produces an array with N elements... integers
x
```

```python
x = np.linspace(1,10,N) # let's put it back to a floating point
x
```

```python
x[1:]
```

```python
x[:-1]
```

```python
x[2:5]
```

```python
dx = x[1:]-x[:-1] # note... this will produce an array with 9 points made up of the difference x[i+1]-x[i]
dx
```

```python
# a basic for loop structure 

for i in range(0,N): # note... the loop will stop one index before N since i would be out of range if i = N
    print(i)
```

```python
ilast = N-1 # the last index of an array of size N
ilast
```

```python
x[ilast] # returns the last element in the array x
```

```python
x[N] # will not return anything since there is no "N" index
```

```python
# basic if statement structure

# runs through the array "x" and prints the index associated with x = 5

for i in range(0,N):
    if x[i] == 5:
        print(i)
```

```python
# basic plotting

x = np.linspace(0,6*np.pi,100)
y = np.cos(x)
y2 = np.cos(x+np.pi/2)

fig,ax = plt.subplots()
ax.plot(x,y, label='cos(x)')
ax.plot(x,y2, label ='cos(x+pi/2)')
ax.set_xlabel('x')
ax.set_ylabel('y');
ax.legend(loc=1);
```

```python
# more basic plotting with some LaTex and figure saving

fig,ax = plt.subplots()
ax.plot(x,np.cos(x), '-r', label='$\cos(x)$')
ax.plot(x,np.cos(x+np.pi/2), '-ok', label ='$\cos(x+\pi/2)$')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$');
ax.legend(loc=4)
plt.savefig('test.pdf');
```
